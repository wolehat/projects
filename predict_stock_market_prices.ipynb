```
#importing the libraries 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
y = stock['S&P 500 Price']
```

```
y
```

```
X = stock.drop(['S&P 500 Price'], axis =1)
```

```
X.head()
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2)
```

```
X_train.shape
```

```
y_train.shape
```

```
X_test.shape
```

```
y_test.shape
```

```
X_train.head()
```

```
y_test.head()
```

```
#model training and evaluation 
from sklearn.linear_model import LinearRegression
```

```
regressor = LinearRegression(fit_intercept=True)
```

```
regressor.fit(X_train, y_train)
```

```
print('Linear Model coefficients (m)', regressor.coef_)  #coef for independent variables 
print('Linear Model coefficients (b)', regressor.intercept_)
```

```
#evaluating the model 
y_predict = regressor.predict(X_test)
```

```
y_predict
```

```
y_test
```

```
plt.scatter(y_test, y_predict, color = 'g')
plt.ylabel('model predictions')
plt.xlabel('True values (ground truth)')
plt.title('Multiple Linear regression prediction')
```

```
#checking the metrics 
k = X_test.shape[1]
n= len(X_test)
```

```
n 
```

```
k
```

```
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from math import sqrt
RMSE =float(format(np.sqrt(mean_squared_error(y_test, y_predict)), '.3f'))
MSE = mean_squared_error(y_test, y_predict)
MAE = mean_absolute_error(y_test, y_predict)
r2 =r2_score(y_test, y_predict)
adj_r2 = 1-(1-r2) * (n-1)/(n-k-1)
MAPE = np.mean(np.abs((y_test-y_predict)/y_test))*100
```

```
print('RMSE =', RMSE)
```

```
print('RMSE =', RMSE, '\nMSE =', MSE, '\nMAE =',MAE, '\nr2 =', r2, '\nAdjusted R2=', adj_r2, '\nMAPE=', MAPE  )
```
